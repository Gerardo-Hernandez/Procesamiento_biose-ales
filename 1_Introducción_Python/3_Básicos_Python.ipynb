# Conceptos básicos de programación

**Variables**: una variable es una ubicación de almacenamiento con nombre que contiene un valor, que se puede cambiar a medida que se ejecuta el programa.

```python
# Variables
x = 10
y = 20
z = x + y
print("z =", z)
```

**Tipos de datos**: diferentes tipos de datos, como números enteros, números de coma flotante, cadenas y valores booleanos, se pueden almacenar en variables y manipular de diferentes maneras.

```python
# Data Types
name = "John"
age = 30
is_male = True
print("Name:", name)
print("Age:", age)
print("Is Male:", is_male)
```

 **Operadores** : Los operadores son símbolos que realizan operaciones sobre valores y variables, como suma, resta, multiplicación y división.

```python
a = 10
b = 5
c = a + b
d = a - b
e = a * b
f = a / b
g = a ** b
print("c =", c)
print("d =", d)
print("e =", e)
print("f =", f)
print("g =", g)
```

**Declaraciones condicionales**: se utilizan para tomar decisiones en función de los valores de las variables. Por ejemplo, se puede usar una declaración if para determinar si un valor es mayor, menor o igual que otro valor.


```python
# Conditional Statements
mark = 75
if mark >= 60:
    print("Pass")
else:
    print("Fail")
```

**Bucles**: los bucles permiten que un bloque de código se ejecute repetidamente en función de una condición. Por ejemplo, se puede usar un bucle for para iterar sobre una lista de valores.


```python
numbers = [10, 20, 30, 40, 50]
for num in numbers:
    print(num)
```

**Funciones**: las funciones son bloques de código reutilizable que se pueden llamar varias veces desde diferentes partes de un programa. Toman entradas (argumentos) y devuelven salidas (valores de retorno).


```python
# Functions
def add(a, b):
    return a + b

result = add(10, 20)
print("Result:", result)
```

**Matrices**: las matrices son estructuras de datos que almacenan colecciones de valores y se pueden manipular mediante bucles y funciones.


```python
# Arrays
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

**Entrada/Salida**: Los programas pueden recibir entradas del usuario o de un archivo y producir salidas que se pueden mostrar en la pantalla o guardar en un archivo.


```python
# Input/Output
name = input("Enter your name:")
print("Hello", name)
```

 **Depuración**: La depuración es el proceso de identificar y corregir errores (bugs) en un programa. Esto puede implicar el uso de herramientas como declaraciones de impresión y el depurador para encontrar y corregir problemas en el código.


```python
# Debugging
# Using print statements to debug code
x = 10
y = 20
z = x + y
print("z =", z)
```

# Cómo generamos una señal en python

El ejemplo más sencillo es generar ondas de la función trignométrica: seno.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sine wave signal
t = np.linspace(0, 2*np.pi, 1000)
y = np.sin(2*np.pi*t)

# Plot the sine wave signal
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.title('Sine Wave Signal')
plt.show()
```

- Este código usa la biblioteca **NumPy** para generar una señal de onda sinusoidal con 1000 puntos en un rango de tiempo de **0 a 2π **segundos.

- La señal se traza utilizando la biblioteca Matplotlib. 

- El eje x de la gráfica representa el tiempo y el eje y representa la amplitud de la señal.

- Este es solo un ejemplo simple, pero se pueden usar técnicas similares para procesar y analizar otros tipos de señales, como audio, video u otras señales de sensores e instrumentos. Mediante el uso de las bibliotecas y técnicas adecuadas, puede realizar operaciones como filtrado, análisis espectral, compresión y más en las señales.

# Cómo cargamos una señal fisiológica en Python

Hay varias bibliotecas utilizadas para leer señales fisiológicas en Python, estas son algunas de las más utilizadas:

- **scipy**: esta biblioteca proporciona una amplia gama de funciones y algoritmos de procesamiento de señales, incluido el soporte para leer y escribir varios formatos de archivo, como WAV, EDF y otros.

- **pyedflib**: esta es una biblioteca diseñada específicamente para leer archivos EDF (formato de datos europeo), que se usan comúnmente en el registro de señales fisiológicas.


> EDF significa European Data Format, un formato de datos para datos de EEG, publicado por primera vez en 1992. En 2003, se publicó una versión mejorada del protocolo de archivo denominado EDF+.

- **biosppy**: esta es una biblioteca para el procesamiento de bioseñales en Python, que incluye soporte para leer y procesar señales de varias fuentes, incluidos ECG, EEG y otros.

> biossppy una caja de herramientas para el procesamiento de bioseñales escrita en Python. La caja de herramientas reúne varios métodos de procesamiento de señales y reconocimiento de patrones orientados al análisis de bioseñales.

> Herramientas:

	> - Soporte para varias bioseñales: BVP, ECG, EDA, EEG, EMG, Respiración
	> - Primitivas de análisis de señales: filtrado, análisis de frecuencia
	> - Agrupación
	> - Biometría

- **neurokit**: esta es una biblioteca para analizar y procesar señales fisiológicas, incluidos ECG, EEG y otras señales. Incluye soporte para leer y escribir varios formatos de archivo, incluidos EDF y otros.

> NeuroKit2 es un paquete fácil de usar que proporciona fácil acceso a rutinas avanzadas de procesamiento de bioseñales. Los investigadores y médicos sin amplios conocimientos de programación o procesamiento de señales biomédicas pueden analizar datos fisiológicos con sólo dos líneas de código .

Estos son solo algunos ejemplos de las bibliotecas disponibles para leer señales fisiológicas en Python. Según sus requisitos específicos y el tipo de señales con las que se desee trabajar, es posible que se deban utilizar una o más de estas bibliotecas, o puede que se requiera escribir un código personalizado para leer y procesar las señales.

# Ahora veamos cómo se aplican

#### Scipy

```python
import scipy.io.wavfile as wav
import numpy as np
import matplotlib.pyplot as plt

# Load a WAV file
filename = '../signals/signal.wav'
fs, data = wav.read(filename)

# Plot the signal data
time = np.arange(len(data)) / fs
plt.plot(time, data)
plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.title('Signal Data')
plt.show()
```

#### pyedflib

```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt

# Load an EDF file
filename = '../signals/signal.edf'
f = pyedflib.EdfReader(filename)
data = f.readSignal(0)

# Plot the signal data
time = np.arange(len(data)) / f.getSampleFrequency(0)
plt.plot(time, data)
plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.title('Signal Data')
plt.show()
```
